## 简单显示 demo

输入对应的 x 值和 y 值，就会进行最原始的图形绘制——折线，如下：

````python
import matplotlib.pyplot as plt  # 注意，是里面的一个子库 pyplot

plt.plot([1, 4, 7, 10], [-2, -4, 8, 5])

plt.show()
````

## numpy 类型数据可视化

### - 单条曲线

````python
import numpy as np

x = np.linspace(-np.pi, np.pi, 100)  # 定义 自变量 取值空间情况
# 是从 -3.141 ~ 3.141 的范围里面，分割100份
plt.plot(x, np.sin(x))

plt.show()
````

###  - 多条曲线

````python
plt.figure(1, dpi=50) # 创建图片，就是整体图形的基础质量？

x_new = np.linspace(-2*np.pi, 2*np.pi, 100)
for i in range(1, 5):
    plt.plot(x_new, np.sin(x_new/i))
    # 如果把 plt.show() 写在里面，那么就会发生一条曲线一个图片的情况
plt.show()
````

### - 直方图

条用方法：```hist()```，如下：

````python
plt.figure(2, dpi=100)
dataHist = [1, 2, 6, 3, 6, 4, 2, 3, 5, 2, 1, 8]
plt.hist(dataHist)  # 方法会自动统计相同数据的“频次”，从而进行绘图

plt.show
````

### - 散点图

方法是：```scatter()```

#### -- 最原始、简单的

````python
x = np.arange(1, 10)
y = 5 / (1+x)
fig = plt.figure()
plt.scatter(x, y, c='b', marker='o')  # c='r' 为 颜色=红色。 marker='o' 指定三点多形状=圆形

plt.show
````

## Pandas 数据绘制

- 1.载入数据

````python
import pandas as pd

iris = pd.read_csv('./iris_training.csv')  # 读取一个文件，可在
# https://github.com/wilsonyin123/geekbangpython/blob/master/python_demo/55/iris_training.csv
# 下载
print(iris.head())  # 显示前五行信息
````

- 2.基于标签是“120”、“4”的两个数据进行绘制

````python
iris.plot(kind='scatter', x='120', y='4')

plt.show()
````

- 3.使用 seaborn 这个库进行绘制

界面更加美观。?

````python
import seaborn as sns

# 指定样子
sns.set(style='white', color_codes=True)

# 普通显示
# sns.jointplot(x='120', y='4', data=iris, size=5)  # .joinplot 就是散点图的方法
# 增加角色的改进显示1
sns.FacetGrid(iris, hue='virginica', size=5).map(plt.scatter,'120','4')
# 单独显示某些列的信息。
sns.FacetGrid(iris, hue='virginica',size=5).map(plt.scatter, 'setosa', 'versicolor').add_legend()
# sns.distplot(iris['120'])
plt.show()
````
