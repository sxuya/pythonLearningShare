### - 获取 baidu.com 的网页的内容

#### -- 建立网址 url

````python
from urllib import request

url = 'http://www.baidu.com'  # 需要告知程序使用的是 http 的协议。
````

#### -- 读取内容

````python
response = request.urlopen(url, timeout=1)  # timeout=1 如果网页打开超过1s，那么就放弃这个操作。（避免网速很卡、很慢，使得程序停滞的状态发生）
# 上面的 urlopen 说明是一个打开操作，内容需要读取出来，即 read
# 直接进行读取看看是什么内容
print(response.read())
````

【说明】  
为什么会和平时我们看到的内容不一样呢？有很多奇怪的字符，比如“...\x90\9c...”？  
因为网页编写的时候，使用的是**“utf-8”**的编码形式。

【背景介绍】  
1. 英文中没有编码这个概念，英文字符，一个字母就站一个位置；
1. 中文的常用子，使用2个位置表示一个汉字的“gbk”编码方式；
1. 包括生僻字，使用的是3个位置表示一个汉字的“utf-8”的编码方式；
1. 如果直接 read，那么程序是按照一个位置一个位置读取，无法解析中文字；
1. so，在 read() 的时候，需要告知程序编码的方式是哪一种。

改进如下：

````python
response = request.urlopen(url, timeout=1)
print(response.read().decode('utf-8'))
````

【说明】  
此时，就能够显示含有中文的正常代码了。

【解释】  
显示的代码，是 HTML 的语言内容，是网页编写使用的语言，即网页的源代码。

### - http 头部信息

#### -- 问题背景  
有时候，我们在用代码进行网页内容获取的时候，会遇到“您的请求被拒绝”，从而导致无法运行。

#### -- 问题原因  
网址的服务器为了避免恶意请求，会进行请求的验证，一个主要的方式就是验证“是否为正规浏览器”，这个信息包含在“http 的头部信息”（“header in http”）。  
如果发现并不是一个浏览器进行数据请求，则会拒绝请求、不会返回我们想要的内容。

#### -- 问题解决  
将代码发起的请求，伪装成一个浏览器，即认为修改 http 的头部信息，使得服务器判定我们是一个浏览器进行访问。

#### -- 代码如下

##### --- 查看代码的头部信息

可以从上面的“代码演示”中的“get”或者“post”部分的返回结果中可以找到，此时的 http 的头部信息是；
>  "headers": {
    ... 
    "User-Agent": "Python-urllib/3.5"
  }, 
  
也就是说，服务器会判定这个请求是 “‘头部’：｛... ‘客户端’：‘Python-urllib/'}” 发起的，而不是一个浏览器。

##### --- 查看浏览器的头部信息

可以在上述“get 文字说明“-“进行测试”中可以看到，浏览器发起请求的 http 头部信息是：  
> "headers": { "Accept": ..., "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/72.0.3626.121 Safari/537.36" },...

可以看到两者的头部信息是不一样的。

##### --- 伪装 http 头部  
自己在使用代码进行内容请求的时候，除了提交 url，还可以一起提交一份自己伪装后的 http 头部。

````python
from urllib import request, parse

url_i = 'http://httpbin.org/post'

header_i = {  # 直接复制 httpbing.org 的 header 内容
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,/;q=0.8", 
#     "Accept-Encoding": "gzip, deflate, br", 
#     为什么加入上面一句话就会报错 UnicodeDecodeError，去掉就好了呢？见最下面获得解析。
    "Accept-Language": "zh-CN,zh;q=0.9,en;q=0.8", 
#     "Connection":"close",
#     "Cookie": "_gauges_unique_hour=1; _gauges_unique_day=1; _gauges_unique_month=1; _gauges_unique_year=1, gauges"
    "Dnt": "1", "Host": "httpbin.org", 
    "Upgrade-Insecure-Requests": "1", 
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/72.0.3626.121 Safari/537.36"
}

dict = {'name':'value'}  # 这里的内容是具体上传数据，这里不做特殊化，只用一般的字符代替。

data_i = bytes(parse.urlencode(dict), encoding='utf8')  # 字典形式的数据无法上传到服务器，要经过加工处理变成 比特 字符才可以。
req = request.Request(url=url_i, data=data_i, headers=header_i, method='POST')
response_i = request.urlopen(req)
print(response_i.read().decode('utf-8'))
````

【结果分析】  
成功伪装！  
虽然看着模型其妙，不就是字符变了嘛，可能现在水平太低，不知道这个到底有什么作用。

【问题解释】  
> 为什么去掉 Accept-Encoding”:'gzip,deflate, br'

首先来看Accept-Encoding的解释： HTTP Header中Accept-Encoding 是浏览器发给服务器,声明浏览器支持的编码类型 （来源于百度百科）；  
说人话就是定义了：客户端和浏览器传输传输”文本内容“时是否需要压缩，而gzip, deflate就是客户端和服务端通用的压缩算法。  
那么为什么会出现上面的UnicodeDecodeError的错误呢？是因为Python默认按照utf-8的编码读取网页文件时，发现是乱码（因为被压缩过了），所以报错了。  
就像是一个txt的文本文件如果被rar压缩软件压缩过，再用记事本打开是乱码是同样的道理。  

所以结论就是要根据服务端的网页编码确定是否需要进行 'Accept-Encoding':' gzip, deflate' 的解压缩操作。
